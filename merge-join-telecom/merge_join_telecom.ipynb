```python
import pandas as pd
```

```python
train = pd.read_csv('/content/train.csv')
event =  pd.read_csv('/content/event_type.csv')
severity = pd.read_csv('/content/severity_type.csv')
resource = pd.read_csv('/content/severity_type.csv')
log = pd.read_csv('/content/log_feature.csv')
```

```python
severity
```

```python
log
```

###1 Merge Tables

```python
merged = train.merge(event)
merged
```

```python
merged2 = merged.merge(severity)
merged2
```

```python
merged3 = merged2.merge(resource)
merged3
```

```python
merged_final = merged3.merge(log)
merged_final
```

###2 Divide dataset into two dataframes.

```python
df_1 = merged_final.iloc[:1000,]
df_2 = merged_final.iloc[1000:,]
```

```python
df_1.tail()
```

```python
df_2
```

###3 Concat the two dataframes.

```python
df = pd.concat([df_1,df_2])
```

###4 Handle duplicates.

```python
nodupe = df.drop_duplicates(subset=['id'])
nodupe
```
